```python
fmri=sns.load_dataset('fmri')
fmri.head()
fmri.tail()
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri) # shade is because there is more than one value for a point.
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci = 'sd')
```

```python
sns.relplot(x = 'timepoint', y = 'signal', estimator = None, kind = 'line', data = fmri) # estimator shows all value in zig zag manner
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'event', kind = 'line', data = fmri)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'region', style = 'event', kind = 'line', data = fmri)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'region', style = 'event', kind = 'line', data = fmri,markers=True,dashes=False)
# The the difference are on the marker or representation only.
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'event', style = 'event', kind = 'line', data = fmri)
# if we give or set same column in hue and style
```
